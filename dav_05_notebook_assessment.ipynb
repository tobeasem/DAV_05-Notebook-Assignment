**Question 4.** Relationships between variables. A scatterplot illustrates a linear relationship between two variables. Consider the  relationship between the dependent variables, `Y1` and `Y2,` and any of the other explanatory variables. Discuss what the relationships you observe. Plot appropriate visualisations and label your plots (title, axes labels, etc) with good descriptions.

```python
plt.figure(figsize=(10,8))
ay1 = plt.subplot(1,2,1)  # 1 by 2 grid, first grid
ay2 = plt.subplot(1,2,2) # 1 by 2 grid, second grid
plt.scatter(data['X1'], data['Y1'], s=10, c='blue', marker='.', alpha=0.6,  ax = ay1, xlabel = 'Relative Compactness-X1', ylabel ='Heating Load-Y1', title ='Scatter Diagram - Relative Compactness vs Heating Load') 
plt.scatter(data['X1'], data['Y2'], s=10, c='blue', marker='.', alpha=0.6), ax = ay2, xlabel = 'Relative Compactness-X1', ylabel ='Cooling Load - Y2', title ='Scatter Diagram - Relative Compactness vs Cooling Load')
```
